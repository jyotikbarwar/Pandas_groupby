```python
#find virat kohli record against all teams
```

```python
virat=df[df["batsman"]=="V Kohli"]
virat.groupby("bowling_team")["batsman_runs"].sum().reset_index()
```

```python
temp_df=df[df["batsman"]=="V Kohli"]
temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).values[0]
```

```python
def max_score(batsman):
    temp_df=df[df["batsman"]==batsman]
    return temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).values[0]
    
```

```python
max_score("CH Gayle")
```
